```python
import sys
from tkinter import *

root = Tk( )

############################## Add Save Tag And Open the Next Window Function ################################################ 
def saveAndExit(): #Creates the Save and Continue Command to save the tags then open the next window
    root.destroy()
##############################################################################################################################

def exit(): #Creates the exit Command to leave the window
    root.destroy()
    
    
root.title("Library Interface") #Creates a basic window
frame = Frame(root)
frame.grid()

scrollbar = Scrollbar(root)
scrollbar.grid(column = 4)

title = Label(root, text="Add Item to Library") #Identifies GUI and Purpose
title.grid(row = 0, column = 1, columnspan = 2)


#Creates all blankspace in the window
blankLbl1 = Label(root, text="")
blankLbl1.grid(row = 6, column = 0)
blankLbl2 = Label(root, text="")
blankLbl2.grid(row = 9, column = 0)
blankLbl3 = Label(root, text="")
blankLbl3.grid(row = 13, column = 0)
blankLbl4 = Label(root, text="")
blankLbl4.grid(row = 19, column = 0)


#Creates Origin Option

originLbl = Label(root, text="Origin: ") #Identifies Origin Option
originLbl.grid(row = 1, column = 0)

originChoices = ("Field Measurement", "b", "c") #Creates the options for Origin

originOpm = OptionMenu(root, StringVar(root), *originChoices) #Creates Origin Dropdown
originOpm.grid(row = 1, column = 3)


#Creates Category Option

catLbl = Label(root, text="Category: ") #Identifies Category Option
catLbl.grid(row = 2, column = 0)

catChoices = ("a", "b", "c") #Creates the options for Category

catOpm = OptionMenu(root, StringVar(root), *catChoices) #Creates Category Dropdown
catOpm.grid(row = 2, column = 3)


#Creates Class Option

classLbl = Label(root, text="Class: ") #Identifies Class Option
classLbl.grid(row = 3, column = 0)
classChoices = ("a", "b", "c") #Creates the options for Class

classOpm = OptionMenu(root, StringVar(root), *classChoices) #Creates Class Dropdown
classOpm.grid(row = 3, column = 3)


#Creates Common Name Option

commonLbl = Label(root, text="Common Name: ") #Identifies Common Name Option
commonLbl.grid(row = 4, column = 0)

commonChoices = ("a", "b", "c") #Creates the options for Common Name

commonOpm = OptionMenu(root, StringVar(root), *commonChoices) #Creates Common Name Dropdown
commonOpm.grid(row = 4, column = 3)


#Creates Name Option

nameLbl = Label(root, text="Name: ") #Identifies Name Option
nameLbl.grid(row = 5, column = 0)

nameEnt = Entry(root) #Creates Name Entry
nameEnt.grid(row = 5, column = 3)


#Creates Mark/Mod Option

mAndMLbl = Label(root, text="Mark/Mod: ") #Identifies Mark/Mod Option
mAndMLbl.grid(row = 7, column = 0)

mAndMChoices = ("a", "b", "c") #Creates the options for Mark/Mod Name

mAndMOpm = OptionMenu(root, StringVar(root), *mAndMChoices) #Creates Mark/Mod Dropdown
mAndMOpm.grid(row = 7, column = 3)


#Creates Serial Number Option

serialLbl = Label(root, text="Serial Number: ") #Identifies Serial Number Option
serialLbl.grid(row = 8, column = 0)

serialEnt = Entry(root) #Creates Serial Number Entry
serialEnt.grid(row = 8, column = 3)


#Creates Orientation Option

orientLbl = Label(root, text="Orientation: ") #Identifies Orientation Option
orientLbl.grid(row = 10, column = 0)

orientChoices = ("Up", "Down", "Left", "Right") #Creates the options for Orientation Name

orientOpm = OptionMenu(root, StringVar(root), *orientChoices) #Creates Orientation Dropdown
orientOpm.grid(row = 10, column = 3)


#Creates Outside Diameter Option

diameterLbl = Label(root, text="Outside Diameter (mm): ") #Identifies Outside Diameter Option
diameterLbl.grid(row = 11, column = 0)

diameterEnt = Entry(root) #Creates Outside Diameter Entry
diameterEnt.grid(row = 11, column = 3)


#Creates Length Option

lengthLbl = Label(root, text="Length (mm): ") #Identifies Length Option
lengthLbl.grid(row = 12, column = 0)

lengthEnt = Entry(root) #Creates Length Entry
lengthEnt.grid(row = 12, column = 3)


#Creates Condition Option

conditionLbl = Label(root, text="Condition: ") #Identifies Condition Option
conditionLbl.grid(row = 14, column = 0)
conditionChoices = ("a", "b", "c") #Creates the options for Condition

conditionOpm = OptionMenu(root, StringVar(root), *conditionChoices) #Creates Condition Dropdown
conditionOpm.grid(row = 14, column = 3)


#Creates Rotating Band Option

bandLbl = Label(root, text="Rotating Band: ") #Identifies Rotating Band Option
bandLbl.grid(row = 15, column = 0)
bandChoices = ("a", "b", "c") #Creates the options for Rotating Band

bandOpm = OptionMenu(root, StringVar(root), *bandChoices) #Creates Rotating Band Dropdown
bandOpm.grid(row = 15, column = 3)


#Creates Fins Option

finsLbl = Label(root, text="Fins: ") #Identifies Fins Option
finsLbl.grid(row = 16, column = 0)
finsChoices = ("a", "b", "c", "d", "e", "f") #Creates the options for Fins

finsOpm = OptionMenu(root, StringVar(root), *finsChoices) #Creates Fins Dropdown
finsOpm.grid(row = 16, column = 3)


#Creates Fuse Option

fuseLbl = Label(root, text="Fuse: ") #Identifies Fuse Option
fuseLbl.grid(row = 17, column = 0)
fuseChoices = ("a", "b", "c") #Creates the options for Fuse

fuseOpm = OptionMenu(root, StringVar(root), *fuseChoices) #Creates Fuse Dropdown
fuseOpm.grid(row = 17, column = 3)


#Creates Spotting Charge Option

chargeLbl = Label(root, text="Spotting Charge: ") #Identifies Spotting Charge Option
chargeLbl.grid(row = 18, column = 0)
chargeChoices = ("a", "b", "c") #Creates the options for Spotting Charge

chargeOpm = OptionMenu(root, StringVar(root), *chargeChoices) #Creates Spotting Charge Dropdown
chargeOpm.grid(row = 18, column = 3)


#Creates Qualifier/Pedigree Option

qualifierLbl = Label(root, text="Qualifier/Pedigree: ") #Identifies Qualifier/Pedigree Option
qualifierLbl.grid(row = 20, column = 0)
qualifierChoices = ("a", "b", "c") #Creates the options for Qualifier/Pedigree

qualifierOpm = OptionMenu(root, StringVar(root), *qualifierChoices) #Creates Qualifier/Pedigree Dropdown
qualifierOpm.grid(row = 20, column = 3)


#Creates Buttons

aplBtn = Button(root, text="Apply", command=saveAndExit) #Creates Apply Button
aplBtn.grid(row = 21, column = 1)

canBtn = Button(root, text="Cancel", command=exit) #Creates Cancel Button
canBtn.grid(row = 21, column = 2)

root.mainloop()
```
